**A) Escriba un código para hacer la gráfica de la densidad de energía por unidad de frecuencia, como

K.**

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.constants as sp
from sklearn.metrics import mean_squared_error
from scipy.integrate import simps

```

```python
#Definición de la función de densidad

def density(wavelength, T): 
    h = 6.62607015e-34
    c = 3.0e8 
    k = 1.380649e-23  
    w = wavelength*10**-9 #Ponemos la escala en (nm)
    u = ((8*np.pi*h)/(w**3))*(1/(np.exp((h*c)/(w*k*T))-1))
    return u  
```

```python
# Valores de longitud de onda
wavelengths = np.linspace(1, 3000, 3000) 

# Valores de temperatura
T1 = 10000
T2 = 8000
T3 = 5000
T4 = 3000


# Gráfica
plt.figure(figsize=(10, 6))
plt.plot(wavelengths, density(wavelengths, T1), label=f'T = {T1} K')
plt.plot(wavelengths, density(wavelengths, T2), label=f'T = {T2} K')
plt.plot(wavelengths, density(wavelengths, T3), label=f'T = {T3} K')
plt.plot(wavelengths, density(wavelengths, T4), label=f'T = {T4} K')
plt.xlabel('Longitud de Onda ($\lambda$) (nm)')
plt.ylabel('Densidad de Energía por unidad de frecuencia ($\mu J/m^3 e15 /GHz$)')
plt.title('Densidad en función de la Longitud de Onda' )
plt.legend()
plt.show()
```

**(b) Calcule la longitud de onda en el pico de cada curva del punto anterior y determine a que parte del

```python
def peak_wavelength(T):
    wavetemp = 0
    dentemp = 0 
    for wavelength in wavelengths:
        if density(wavelength, T) > dentemp :
            wavetemp = wavelength
            dentemp = density(wavelength, T)

    return wavetemp, dentemp

print(peak_wavelength(T1))
```

```python
plt.figure(figsize=(10, 6))
plt.plot(wavelengths, density(wavelengths, T1), label=f'T = {T1} K')
plt.plot(wavelengths, density(wavelengths, T2), label=f'T = {T2} K')
plt.plot(wavelengths, density(wavelengths, T3), label=f'T = {T3} K')
plt.plot(wavelengths, density(wavelengths, T4), label=f'T = {T4} K')

plt.scatter(peak_wavelength(T1)[0], peak_wavelength(T1)[1], label=f'$\lambda$ = {peak_wavelength(T1)[0].round(2)} nm')
plt.scatter(peak_wavelength(T2)[0], peak_wavelength(T2)[1], label=f'$\lambda$ = {peak_wavelength(T2)[0].round(2)} nm')
plt.scatter(peak_wavelength(T3)[0], peak_wavelength(T3)[1], label=f'$\lambda$ = {peak_wavelength(T3)[0].round(2)} nm')
plt.scatter(peak_wavelength(T4)[0], peak_wavelength(T4)[1], label=f'$\lambda$ = {peak_wavelength(T4)[0].round(2)} nm')
plt.xlabel('Longitud de Onda ($\lambda$) (nm)')
plt.ylabel('Densidad de Energía por unidad de frecuencia ($\mu J/m^3  /GHz$)')
plt.title('Densidad en función de la Longitud de Onda' )
plt.legend()
plt.show()
```

**(c) Grafíque una de las curvas del punto 1 como función de la frecuencia. En la misma gráfica, incluya
la curva de Rayleigh-Jeans, para bajas frecuencias. Haga una comparación cuantitativa del nivel
de concordancia entre las dos curvas. Repita el mismo ejercicio, pero ahora para la predicción de
Wien. Tiene libertad para medir la concordancia entre las curvas.**

```python
h = 6.626e-34
c = 3.0e8 
k = 1.38e-23 
    
def planck(frec, T):  
    frec = frec*10**12
    u = ((8*np.pi*h*frec**3)/(c**3))*(1/(np.exp((h*frec)/(k*T))-1)) 
    return u

def rayleigh_jeans(frec, T):
    f = frec*10**12
    u = (8*np.pi*f**2/c**3)*k*T
    return u


def weinn(frec, T):
    frec = frec*10**12
    u = ((8*np.pi*h*frec**3)/c**3)*(np.exp(-h*frec/(k*T))) 
    
    return u
```

```python
# Valores de longitud de onda
frecu = np.linspace(1, 1000, 1000) 

# Valores de temperatura
T6 = 5000

# Gráfica
plt.figure(figsize=(10, 6))
plt.plot(frecu, planck(frecu, T6), label=f'Planck')
plt.plot(frecu, rayleigh_jeans(frecu, T6), label=f'Rayleigh_Jeans')
plt.plot(frecu, weinn(frecu, T6), label=f'Wien')
plt.xlabel('Frecuencia (THz)')
plt.ylabel('Densidad de Energía por unidad de frecuencia ($\mu J/m^3 e15 /GHz$ )')
plt.title('Densidad en función de la frecuencia para T = 5000 K' )
plt.ylim(0, 1.5*10**(-15))
plt.legend()
plt.show()
```

```python

# Medir concordancia
curve_planck = planck(frecu, T6)
curve_rayleigh = rayleigh_jeans(frecu, T6)

mse = mean_squared_error(curve_planck, curve_rayleigh)
abc = simps(np.abs(curve_planck - curve_rayleigh))
print("Área entre Curvas:", abc)
print("Error Cuadrático Medio:", mse)


frecus = np.linspace(1, 500, 500) 
def error():
    
    distances = np.array([])
    for i in frecus:
        d =  100*abs(planck(i, T6)-rayleigh_jeans(i, T6))/ planck(i, T6)
        distances = np.append(distances, d)
        
    return distances

peak_error = 100*abs(planck(293.829578844270, T6)-rayleigh_jeans(293.829578844270, T6))/ planck(293.829578844270, T6)
print("Error porcentual del pico de frecuencia: ", peak_error, "%")
print(c/(1021*10**-9))

plt.plot(frecus, error(), label=f'Planck')

plt.title("Error porcentual de la aproximación de Rayleigh Jeans con respecto a la Curva de Planck")
plt.ylabel("Error Procentual % ")
plt.xlabel("frecuencia (THz)")
plt.show
```

Para observar la concordancia entre las curvas se calculó el error porcentual de la curva de Rayleigh Jeans con respecto a la curva de Planck para cada frecuencia antes del pico. El resultado se muestra en la gráfica de arriba. También, se calculó que el error porcentual para el pico es de 460%. Se observa además que las aproximaciones son más cercanas para frecuencias bajas y crecen, aparentemente de manera exponencial a medida que se incrementa la frecuencia.
